#### (1.b)

```python
def grade(stuple):
    
    for i in stuple: 
        
        n = (stuple[0]*0.35)+(stuple[1]*0.35)+(stuple[2]*0.3)
        
        if (0 <= n < 35):
            
            return('Final grade: F')
            
        elif (35 <= n < 45):
            
            return('Final grade: E')
        
        elif (45 <= n < 55):
            
            return('Final grade: D')
            
        elif (55 <= n < 65):
            
            return('Final grade: C')
        
        elif (65 <= n < 75):
            
            return('Final grade: B')
            
        elif (75 <= n <= 100):
            
            return('Final grade: A')
```

#### (1.c)

```python
b = (70,50,65)
c = (80,75,90)
```

```python
print(grade(b))

print(grade(c))
```

## Part 3

#### (3.a)

```python
ID1 = pd.DataFrame({"CustomerID":["C01","C02","C03","C05","C06","C08","C10"],
                    "Gender":["F","M","M","F","M","F","F"]})
ID1
```

```python
ID2 = pd.DataFrame({"CustomerID":["C01","C03","C04","C06","C07","C08","C09"],
                    "Age":[20,45,25,35,40,30,55]})
ID2
```

#### (3.b)

```python
ID = ID1.merge(ID2, 
               how='outer', 
               on='CustomerID')

ID
```

#### (3.c)

```python
IDU = ID.fillna('Unknown').sort_values(by = 'CustomerID', 
                                       ascending = True).reset_index(drop = True)

IDU
```

#### (3.d)

```python
IDU.groupby('Gender').size().plot(kind = "bar", 
                                  y = "Gender", 
                                  edgecolor = "black")
```
